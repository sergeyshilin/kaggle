```python
%matplotlib inline

import gmplot
import numpy as np
import pandas as pd
import matplotlib
from random import randint
from random import shuffle
from collections import Counter
import matplotlib.pyplot as plt
pylab.rcParams['figure.figsize'] = 16, 12  # that's default image size for this interactive session
```

```python
# Read all the train and test data.
# It is not recommended to recompile this block
train_data = pd.read_csv('data/train.csv', header=False)
test_data = pd.read_csv('data/test.csv', header=False)
```

```python
train_x = np.array(train_data.ix[:,:])
train_x = np.delete(train_x, 1, 1)
train_y = np.array(train_data.ix[:,1])
```

```python
# Count distinct categories of crimes
crimes = Counter(train_y).items()
vals = []
nums = []
for i in range(len(crimes)):
    vals.append(str(crimes[i][0]))
    nums.append(int(crimes[i][1]))
    

fig = pl.figure()
ax = pl.subplot(111)
width=0.8
ax.bar(range(len(vals)), nums, width=width)
ax.set_xticks(np.arange(len(vals)) + width/2)
ax.set_xticklabels(vals, rotation=90)
fig.show()
```

```python
# Prepare array of colors
colors = []

for i in range(len(crimes)):
    colors.append('%06X' % randint(0, 0xFFFFFF))

print colors
```

```python
#Prepare data to display on map
data_to_show = []
for i in range(len(crimes)):
    crime = vals[i]
    crimeindexes = list(numpy.where(train_y==crime)[0])
    crimelatlongs = train_x[crimeindexes][:, -2:]
    data_to_show.append([vals[i], crimelatlongs])
```

```python
def getLatsLongs(data, i, count):
    latslong = data_to_show[i][1].tolist()
    shuffle(latslong)
    latslong = latslong[:count]
    lats = []
    longs = []
    for val in latslong:
        lats.append(val[1])
        longs.append(val[0])
    
    return lats, longs
```

```python
latslong = data_to_show[15][1].tolist()
shuffle(latslong)
latslong = latslong[:1000]
print latslong[0][1]
```

```python
gmap = gmplot.GoogleMapPlotter(37.763116, -122.440224, 13)

for i in range(len(data_to_show)):
    lats, longs = getLatsLongs(data_to_show, i, 1000)
    color = "#" + colors[i]
    gmap.scatter(lats, longs, color, size=40, marker=False)


gmap.draw("crimesByColor.html")
```
