```python
import os
import pandas as pd
import wikipedia as wiki
from urllib2 import urlopen
from bs4 import BeautifulSoup
```

```python
valid_set = pd.read_csv('data/validation_set.tsv', sep = '\t' )
```

```python
valid_set[:3]
```

```python
all_answers = []
all_answers.extend(valid_set['answerA'].values)
all_answers.extend(valid_set['answerB'].values)
all_answers.extend(valid_set['answerC'].values)
all_answers.extend(valid_set['answerD'].values)

unique_answers = set(all_answers)
```

```python
save_folder = 'data/wiki_answers'

for kw in list(unique_answers)[4467:]:
    try:
        content = wiki.page(kw).content.encode('ascii', 'ignore')
    except wiki.exceptions.DisambiguationError as e:
        print 'DisambiguationError', kw
    except:
        print 'Error', kw
    if not content:
        continue
    else:
        with open(os.path.join(save_folder, '_'.join(kw.split()) + '.txt'), 'w') as f:
                f.write(content)
```
