```python
f = open("acid_rain.txt","r")
lines = f.readlines()
f.close()
```

```python
headers_to_remove = [
    '== External links',
    '== References',
    '== See also',
    '== Notes and references',
    '=== Further reading',
    '== Notes',
    '== Additional Resources',
    '== Further reading'
]
```

```python
for st_r in headers_to_remove:
    print st_r
```

```python
temp_headers = headers_to_remove[:]
new_lines = []
for i in xrange(len(lines)-1, 0, -1):
    for st_r in temp_headers:
        if(st_r in lines[i]):
            new_lines = lines[:i]
            temp_headers.remove(st_r)
            break
```

```python
new_lines
```

```python
f = open("acid_rain.txt","w")
for line in new_lines:
    f.write(line)

f.close()
```
